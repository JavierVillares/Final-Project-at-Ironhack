```python
import pandas as pd
```

This is so we can create an easy pie chart of total population and find out how much of the total pop is under 2.1 birth rate

```python
url = ('/Users/javi/Desktop/Ironhack/Entregas_de_github/Final-Project-at-Ironhack/Data/Cleaned/MEGAMERGE_machine_learning_ready.csv')
df = pd.read_csv(url)
df
```

```python
# Filter out rows where 'Country Name' is 'World'
df_filtered = df[df['Country Name'] != 'World']

# Group by year and calculate the required statistics
df_filtered['Above 2.1'] = df_filtered['Birth Rate'] >= 2.1

# Aggregate df
df_agg = df_filtered.groupby(['Year', 'Above 2.1']).agg({
    'Total Population': 'sum'
}).reset_index()

# Calculate the total population per year
total_population_per_year = df_agg.groupby('Year')['Total Population'].sum().reset_index()
total_population_per_year = total_population_per_year.rename(columns={'Total Population': 'Total Population Year'})

# Merge the aggregated df with total population per year
df_merged = pd.merge(df_agg, total_population_per_year, on='Year')

# Calculate the percentage
df_merged['Percentage'] = (df_merged['Total Population'] / df_merged['Total Population Year']) * 100

# Prepare the final df
final_df = df_merged.pivot(index='Year', columns='Above 2.1', values='Percentage').reset_index()
final_df = final_df.rename(columns={False: 'Below 2.1', True: 'Above or Equal 2.1'})


# Save the prepared df to CSV
final_df.to_csv('/Users/javi/Desktop/Ironhack/Entregas_de_github/Final-Project-at-Ironhack/Data/Cleaned/MEGAMERGE_machine_learning_percentages.csv', index=False)
```
