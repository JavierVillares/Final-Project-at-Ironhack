```python
url = ('/Users/javi/Desktop/Ironhack/Entregas_de_github/Final-Project-at-Ironhack/Data/Raw data/World demographic/Gdp per capita/Gpd_per_capita.csv')
df = pd.read_csv(url, sep=';')
```

```python
# Function to check the percentage of NaN values in each row
def check_nan_percentage(row):
    nan_count = row.isna().sum()
    total_count = len(row) - 1  # Exclude the 'Country Name' column
    nan_percentage = nan_count / total_count
    return nan_percentage

# List to store countries with more than 35% NaN values
countries_with_high_nan = []

# Process each row
for index, row in df.iterrows():
    nan_percentage = check_nan_percentage(row)
    if nan_percentage > 0.35:
        countries_with_high_nan.append(row['Country Name'])
    else:
        # Forward fill, backward fill, and interpolate for rows with <= 35% NaN values
        df.iloc[index, 1:] = df.iloc[index, 1:].ffill().bfill().interpolate()

# Display the countries with more than 35% NaN values
if countries_with_high_nan:
    print("Countries with more than 35% NaN values:")
    for country in countries_with_high_nan:
        print(country)
else:
    print("No country has more than 35% NaN values.")
```

```python
def calculate_nan_percentage(row):
    nan_count = row.isna().sum()
    total_count = len(row) - 1  # Exclude the 'Country Name' column
    nan_percentage = (nan_count / total_count) * 100
    return nan_percentage

# Apply the function to each row and create a new column for NaN percentage
df['NaN Percentage'] = df.apply(calculate_nan_percentage, axis=1)

# Display the countries and their respective NaN percentages
nan_percentage_df = df[['Country Name', 'NaN Percentage']]
nan_percentage_df
```

```python
countries_with_nan = nan_percentage_df[nan_percentage_df['NaN Percentage'] > 0]

print(countries_with_nan)
```

```python
def calculate_nan_percentage(row):
    nan_count = row.isna().sum()
    total_count = len(row) - 1  # Exclude the 'Country Name' column
    nan_percentage = (nan_count / total_count) * 100
    return nan_percentage

# Apply the function to each row and create a new column for NaN percentage
df['NaN Percentage'] = df.apply(calculate_nan_percentage, axis=1)

# Filter rows where NaN percentage is greater than 35%
rows_to_drop = df[df['NaN Percentage'] > 35]

# Display the countries with more than 30% NaN values
if not rows_to_drop.empty:
    print("Countries with more than 35% NaN values:")
    print(rows_to_drop[['Country Name', 'NaN Percentage']])
else:
    print("No country has more than 35% NaN values.")

# Drop these rows from the main DataFrame
df_cleaned = df[df['NaN Percentage'] <= 35].drop(columns=['NaN Percentage'])

# Display the cleaned DataFrame
print(df_cleaned)
```

```python
df_long = pd.melt(df_cleaned, id_vars=['Country Name'], var_name='Year', value_name='GDP_per_capita')

# Convert 'Year' to an integer
df_long['Year'] = df_long['Year'].astype(int)

df_sorted = df_long.sort_values(by=['Country Name','Year']).round(decimals=2)

# Reset the index
df_sorted.reset_index(drop=True, inplace=True)
```
