```python
# Get a list of valid country names using pycountry
country_names = [country.name for country in pycountry.countries]

# Filter the datasets to include only rows where 'Country Name' is a valid country name
birth_rate_df = birth_rate_df[birth_rate_df['Country Name'].isin(country_names)]
gdp_per_capita_df = gdp_per_capita_df[gdp_per_capita_df['Country Name'].isin(country_names)]

# Convert 'Year' columns to appropriate data types if necessary
birth_rate_df['Year'] = birth_rate_df['Year'].astype(int)
gdp_per_capita_df['Year'] = gdp_per_capita_df['Year'].astype(int)

# Merge the datasets on 'Country Name' and 'Year'
merged_df = pd.merge(birth_rate_df, gdp_per_capita_df, on=['Country Name', 'Year'])

# Calculate the correlation between GDP per Capita and Birth Rate for each country
correlations = {}
for country in merged_df['Country Name'].unique():
    country_data = merged_df[merged_df['Country Name'] == country]
    if len(country_data) > 1:  # Ensure there's enough data to calculate correlation
        correlation = country_data['GDP_per_capita'].corr(country_data['Birth-rate'])
        correlations[country] = correlation

# Convert the correlations dictionary to a DataFrame for better visualization
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Country Name', 'Correlation'])

# Sort the DataFrame by the absolute value of correlation to find the highest correlations
correlation_df['Abs Correlation'] = correlation_df['Correlation'].abs()
sorted_correlation_df = correlation_df.sort_values(by='Abs Correlation', ascending=False)

# Display the top 10 countries with the highest correlation
print("Top 10 Countries with the Highest Correlation between GDP per Capita and Birth Rate:")
print(sorted_correlation_df.head(50))

# Plot the correlations
plt.figure(figsize=(12, 8))
sns.barplot(x='Correlation', y='Country Name', data=sorted_correlation_df.head(50))
plt.title('Top 10 Countries with the Highest Correlation between GDP per Capita and Birth Rate')
plt.xlabel('Correlation')
plt.ylabel('Country Name')
plt.show()
```
