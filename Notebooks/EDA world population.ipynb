```python
# Choose specific countries for the analysis
countries = ['World', 'United States', 'China', 'India', 'Japan', 'South Korea']

# Filter datasets by the chosen countries
birth_rate_filtered = df[df['Country Name'].isin(countries)]
death_rate_filtered = df[df['Country Name'].isin(countries)]
gdp_per_capita_filtered = df[df['Country Name'].isin(countries)]
population_growth_filtered = df[df['Country Name'].isin(countries)]
median_age_filtered = df[df['Country Name'].isin(countries)]
secondary_enrolment_filtered = df[df['Country Name'].isin(countries)]

# Plot trends for each country
plt.figure(figsize=(14, 15))

# Birth Rate Trend
plt.subplot(3, 2, 1)
for country in countries:
    country_data = birth_rate_filtered[birth_rate_filtered['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Birth Rate'], label=country)
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.title('Birth Rate Over Time')
plt.legend()

# Death Rate Trend
plt.subplot(3, 2, 2)
for country in countries:
    country_data = death_rate_filtered[death_rate_filtered['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Death Rate'], label=country)
plt.xlabel('Year')
plt.ylabel('Death Rate')
plt.title('Death Rate Over Time')
plt.legend()

# GDP per Capita Trend
plt.subplot(3, 2, 3)
for country in countries:
    country_data = gdp_per_capita_filtered[gdp_per_capita_filtered['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['GDP per Capita'], label=country)
plt.xlabel('Year')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita Over Time')
plt.legend()

# Population Growth Rate Trend
plt.subplot(3, 2, 4)
for country in countries:
    country_data = population_growth_filtered[population_growth_filtered['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Natural Growth Rate'], label=country)
plt.xlabel('Year')
plt.ylabel('Population Growth Rate')
plt.title('Population Growth Rate Over Time')
plt.legend()

# Median Age Trend
plt.subplot(3, 2, 5)
for country in countries:
    country_data = median_age_filtered[median_age_filtered['Country Name'] == country]
    plt.plot(country_data['Year'], country_data['Median Age'], label=country)
plt.xlabel('Year')
plt.ylabel('Median Age')
plt.title('Median Age Over Time')
plt.legend()

```

```python
# Extract relevant columns for analysis
birth_rate_df = df[['Country Name', 'Year', 'Birth Rate', 'Natural Growth Rate']]

# Filter out countries that started with a birth rate below 2.1
initial_birth_rate = birth_rate_df.groupby('Country Name').first().reset_index()
countries_above_2_1_initially = initial_birth_rate[initial_birth_rate['Birth Rate'] > 2.1]['Country Name']

# Filter the df for these countries
filtered_data = birth_rate_df[birth_rate_df['Country Name'].isin(countries_above_2_1_initially)]

# Find the year when birth rate fell below 2.1 for each country
below_2_1 = filtered_data[filtered_data['Birth Rate'] < 2.1]
year_below_2_1 = below_2_1.groupby('Country Name').first().reset_index()

# Find the first year of negative growth rate for each country
negative_growth_rate = filtered_data[filtered_data['Natural Growth Rate'] < 0]
year_negative_growth = negative_growth_rate.groupby('Country Name').first().reset_index()

# Merge the two dataframes to calculate the difference in years
merged_data = pd.merge(year_below_2_1, year_negative_growth, on='Country Name', suffixes=('_below_2_1', '_negative_growth'))
merged_data['Years_to_Negative_Growth'] = merged_data['Year_negative_growth'] - merged_data['Year_below_2_1']

# Calculate the average number of years
average_years_to_negative_growth = merged_data['Years_to_Negative_Growth'].mean()
print("The average number of years after a country has sarted to decline in it's growth rate :", average_years_to_negative_growth)
```

```python
# Filter df for the year 2021 and exclude the "World" country
df_2021 = df[(df['Year'] == 2021) & (df['Country Name'] != 'World')]

# Calculate the total population of countries with a birth rate below 2.1
population_below_2_1 = df_2021[df_2021['Birth Rate'] < 2.1]['Total Population'].sum()

# Calculate the total population of all countries
total_population = df_2021['Total Population'].sum()

# Calculate the percentage
percentage_below_2_1 = (population_below_2_1 / total_population) * 100
print("Percentage of total population fallen below replacement rate",round(percentage_below_2_1, 2))
```

```python
# Calculate average education metrics for countries with birth rate below 2.1
education_below_2_1 = df_2021[df_2021['Birth Rate'] < 2.1]
average_tertiary_enrolment_f_below_2_1 = education_below_2_1['Tertiary Enrolment (F)'].mean()
average_tertiary_enrolment_m_below_2_1 = education_below_2_1['Tertiary Enrolment (M)'].mean()

# Calculate average education metrics for countries with birth rate above or equal to 2.1
education_above_2_1 = df_2021[df_2021['Birth Rate'] >= 2.1]
average_tertiary_enrolment_f_above_2_1 = education_above_2_1['Tertiary Enrolment (F)'].mean()
average_tertiary_enrolment_m_above_2_1 = education_above_2_1['Tertiary Enrolment (M)'].mean()

average_education = {
    'Below 2.1 Birth Rate': {
        'Tertiary Enrolment (F)': average_tertiary_enrolment_f_below_2_1,
        'Tertiary Enrolment (M)': average_tertiary_enrolment_m_below_2_1
    },
    'Above 2.1 Birth Rate': {
        'Tertiary Enrolment (F)': average_tertiary_enrolment_f_above_2_1,
        'Tertiary Enrolment (M)': average_tertiary_enrolment_m_above_2_1
    }
}

average_education
```

```python
# Calculate average primary and secondary enrolment for countries with birth rate below 2.1
average_primary_enrolment_f_below_2_1 = education_below_2_1['Primary Enrolment (F)'].mean()
average_primary_enrolment_m_below_2_1 = education_below_2_1['Primary Enrolment (M)'].mean()
average_secondary_enrolment_f_below_2_1 = education_below_2_1['Secondary Enrolment (F)'].mean()
average_secondary_enrolment_m_below_2_1 = education_below_2_1['Secondary Enrolment (M)'].mean()

# Calculate average primary and secondary enrolment for countries with birth rate above or equal to 2.1
average_primary_enrolment_f_above_2_1 = education_above_2_1['Primary Enrolment (F)'].mean()
average_primary_enrolment_m_above_2_1 = education_above_2_1['Primary Enrolment (M)'].mean()
average_secondary_enrolment_f_above_2_1 = education_above_2_1['Secondary Enrolment (F)'].mean()
average_secondary_enrolment_m_above_2_1 = education_above_2_1['Secondary Enrolment (M)'].mean()

average_education.update({
    'Below 2.1 Birth Rate': {
        'Primary Enrolment (F)': average_primary_enrolment_f_below_2_1,
        'Primary Enrolment (M)': average_primary_enrolment_m_below_2_1,
        'Secondary Enrolment (F)': average_secondary_enrolment_f_below_2_1,
        'Secondary Enrolment (M)': average_secondary_enrolment_m_below_2_1,
        'Tertiary Enrolment (F)': average_tertiary_enrolment_f_below_2_1,
        'Tertiary Enrolment (M)': average_tertiary_enrolment_m_below_2_1
    },
    'Above 2.1 Birth Rate': {
        'Primary Enrolment (F)': average_primary_enrolment_f_above_2_1,
        'Primary Enrolment (M)': average_primary_enrolment_m_above_2_1,
        'Secondary Enrolment (F)': average_secondary_enrolment_f_above_2_1,
        'Secondary Enrolment (M)': average_secondary_enrolment_m_above_2_1,
        'Tertiary Enrolment (F)': average_tertiary_enrolment_f_above_2_1,
        'Tertiary Enrolment (M)': average_tertiary_enrolment_m_above_2_1
    }
})

average_education
```
